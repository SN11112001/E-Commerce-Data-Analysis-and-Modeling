```python
X = df.iloc[:,:-1]

X
```

```python
y = df.iloc[:,-1]

y
```

#### a. Performing One Hot Encoding on Categorical Features.

```python
X = pd.get_dummies(data = X, columns = ["Avatar"])

X
```

Splitting data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

<div style="font-family:Trebuchet MS; border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="left"><font color=orange>ðŸ“Š Inference:</font></h3><br>
    
    
1. "One Hot Encoding" is performed on the categorical column "Avatar", adding 5 more columns to the training dataset.<br>
    <br>
3. The dataset is seperated into Dependent Set and Independent Set, and further splitted into 80% training and rest 20% as testing set, using train_test_split.<br>

---

<div style="border-radius:10px; border:black solid; padding: 20px; background-color: #FED8B1; font-size:100%; text-align:left">
<p style="font-family:Trebuchet MS; font-weight:bold; letter-spacing: 0px; color:black; font-size:150%; text-align:center;padding: 0px;">5. Model Creation using "Multiple Linear Regression".</p></div>

#### a. Fitting the Model with data.

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

#### b. Predicting the values with model.

```python
y_pred
```

#### c. Getting the intercept and slope values for the final equation of modal.

```python
regressor.intercept_
```

```python
regressor.coef_
```

#### d. Getting the Equation with the features name.

```python
string = "Annual Amount Spent = "

for i in range(len(X.columns)):
    string = string + "(" + str(round(regressor.coef_[i],2)) + " " + X.columns[i] + ")" + " + "
    
string
```

#### e. Measuring Performace Metrics

```python
score=r2_score(y_test,y_pred)

print("R Squared:", score)
```

```python
mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
```

```python
rmse = mean_squared_error(y_test, y_pred, squared=False)

print("Root Mean Squared Error (RMSE):", rmse)
```

```python
mae = mean_absolute_error(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
```

#### f. Plotting the model predicted values along side actual values.

```python
plt.figure()

plt.scatter(y_test, y_pred)

plt.show()
```
